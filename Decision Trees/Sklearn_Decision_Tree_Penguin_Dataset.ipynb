### Cleaning Data

```python
# Menghapus kolom yang tidak digunakan
penguins.drop(['sex','island'], axis=1, inplace=True)
```

```python
# Mengisi null value dengan mean
mean_cl = penguins['culmen_length_mm'].mean()
mean_cd = penguins['culmen_depth_mm'].mean()
mean_fl = penguins['flipper_length_mm'].mean()
mean_bm = penguins['body_mass_g'].mean()
penguins['culmen_length_mm'].fillna(mean_cl, inplace=True)
penguins['culmen_depth_mm'].fillna(mean_cd, inplace=True)
penguins['flipper_length_mm'].fillna(mean_fl, inplace=True)
penguins['body_mass_g'].fillna(mean_bm, inplace=True)
```

```python
penguins.isna().sum()
```

### Decision Tree step

```python
# memisahkan atribut dan label
x = penguins[['culmen_length_mm','culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = penguins['species']
```

```python
# melakukan normalisasi agar data numerik memiliki skala yang sama
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data, columns= x.columns)
scaled_data.head()
```

```python
# membagi dataset menjadi data train dan data test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=42)
```

```python
# membuat model decision_tree
from sklearn.tree import DecisionTreeClassifier

tree_model=DecisionTreeClassifier()
```

```python
# melakukan pelatihan model terhadap data
tree_model = tree_model.fit(x_train, y_train)
```

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

y_predicted = tree_model.predict(x_test)
akurasi_skor = round(accuracy_score(y_predicted, y_test),4)
print("skor akurasi:", akurasi_skor)
```

```python
# prediksi model dengan tree_model.predict([[culmen, SepalWidth, PetalLength, PetalWidth]])
test_predict = tree_model.predict([[44.5, 19.7, 200.1, 4000]])
print(test_predict[0])
```

```python
#exprt the decision tree graph
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "penguins.dot",
    feature_names = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'],
    class_names = ['Adélie', 'Chinstrap ', 'Gentoo' ],
    rounded= True,
    filled =True
)
```
